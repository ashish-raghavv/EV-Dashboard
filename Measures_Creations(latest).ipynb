```python

# Define functions for calculating single teacher counts
def calculate_single_teacher_students(row, gender, min_students, max_students):
    if gender == 'male':
        # Example calculation logic for male students within 51-100 range
        return row['total_male_students'] if min_students <= row['total_male_students'] <= max_students else 0
    else:
        # Example calculation logic for female students within 51-100 range
        return row['total_female_students'] if min_students <= row['total_female_students'] <= max_students else 0

def calculate_single_teacher_teachers_category(row, gender, min_students, max_students):
    if gender == 'male':
        # Example calculation logic for male teachers within 51-100 range
        return row['total_male_teachers_category'] if min_students <= row['total_male_teachers_category'] <= max_students else 0
    else:
        # Example calculation logic for female teachers within 51-100 range
        return row['total_female_teachers_category'] if min_students <= row['total_female_teachers_category'] <= max_students else 0

# Calculating new columns
df['single_teacher_51_100_male_students'] = df.apply(lambda row: calculate_single_teacher_students(row, 'male', 51, 100), axis=1)
df['single_teacher_51_100_female_students'] = df.apply(lambda row: calculate_single_teacher_students(row, 'female', 51, 100), axis=1)

# Assuming teacher categories for 51-100 student ranges are calculated similarly
df['single_teacher_51_100_male_teachers_category'] = df.apply(lambda row: calculate_single_teacher_teachers_category(row, 'male', 51, 100), axis=1)
df['single_teacher_51_100_female_teachers_category'] = df.apply(lambda row: calculate_single_teacher_teachers_category(row, 'female', 51, 100), axis=1)

# Check results
print(df[['single_teacher_51_100_male_students', 'single_teacher_51_100_female_students', 
          'single_teacher_51_100_male_teachers_category', 'single_teacher_51_100_female_teachers_category']].head())
```
